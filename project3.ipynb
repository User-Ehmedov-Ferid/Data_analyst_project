```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("store.xlsx")
```

```python
data['Postal Code'] =  data['Postal Code'].replace(np.nan , 0)
```

```python
pd.options.display.max_rows = 147
data['Country'].value_counts()
```

```python
data_selected =  data[data['Country'].isin(['United States' , 'Australia' , 'France' , 'Mexico' , 'Germany' , 'China' , 'United Kingdom' , 'Brazil' , 'India' , 'Indonesia' ])]
```

```python
data_selected.head()
```

```python
data_selected.sort_values(by = 'Country' , inplace=True)
```

```python
returns =  pd.read_excel('store.xlsx' , sheet_name= 'returns') 
```

```python
data_selected =  data_selected.merge(returns , how='inner' , on = 'Order ID')
```

```python
data_selected =  data_selected[data_selected['Returned'] == 'Yes']
```

```python
data_selected
```

```python
data_selected['Year'] = data_selected['Ship Date'].astype(str).str[:4]
```

```python
sales =  data_selected.groupby(by = ['Year' ,'Country' ] ).sum().sort_values('Year').reset_index()
```

```python
sales = sales.drop(columns=['Row ID' , 'Postal Code' ])
```

```python
sales.to_excel('New0 Sales.xlsx')
```
