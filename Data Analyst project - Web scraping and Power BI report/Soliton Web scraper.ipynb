```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# Get all urls 

```python
all_urls = []
```

```python
main_url = 'https://www.soliton.az/az/komputer-ve-aksesuarlar/noutbuklar/?sorting=0-9'
headers2 = {"User Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
page1 = requests.get(main_url , headers = headers2)
soup3 = BeautifulSoup(page1.content, 'html.parser')
soup4 = BeautifulSoup(soup3.prettify() , 'html.parser')
all_page = soup4.find_all('div' , {'class' : 'pic'})

for d in all_page:
    d1 = d.find('a')['href']
    all_urls.append("https://www.soliton.az/" + d1)
```

# Get Titles and Prices of Computers

```python
title = soup4.find_all("a" , class_ = "title")
price = soup4.find_all("div" , class_ = "price")
```

```python
title2 = []
for x in title:
    m = x.get_text().replace('\n        ' , '').replace('\n       ' , '')
    title2.append(m)
```

```python
price2 = []
for y in price:
    n = y.get_text().replace('\n\n\n          ' , '').replace('\n         \n\n          M\n         \n\n' , '').replace('\n         \n\n          M\n         \n' , '')
    price2.append(n)
```

# Get features of computers

```python
processor = []
for urls in all_urls:
	page = requests.get(urls , headers = headers2)
	soup1 = BeautifulSoup(page.content , 'html.parser')
	soup2 = BeautifulSoup(soup1.prettify() , 'html.parser')

	feature1 = soup2.find_all('td')[3]
	text = feature1.text.strip()
	processor.append(text)
```

```python
ram = []
for urls in all_urls:
	page = requests.get(urls , headers = headers2)
	soup1 = BeautifulSoup(page.content , 'html.parser')
	soup2 = BeautifulSoup(soup1.prettify() , 'html.parser')

	feature2 = soup2.find_all('td')[5]
	text = feature2.text.strip()
	ram.append(text)
```

```python
disk = []
for urls in all_urls:
	page = requests.get(urls , headers = headers2)
	soup1 = BeautifulSoup(page.content , 'html.parser')
	soup2 = BeautifulSoup(soup1.prettify() , 'html.parser')

	feature3 = soup2.find_all('td')[7]
	text = feature3.text.strip()
	disk.append(text)
```

```python
pd.set_option('display.max_rows', None)
```

```python
data = pd.DataFrame({'Title': pd.Series(title2) ,'Processor': pd.Series(processor) , 'Ram': pd.Series(ram) , 'Disk': pd.Series(disk) , 'Price': pd.Series(price2)})
```

```python
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```

```python
data.to_excel('Soliton Datas2.xlsx')
```
