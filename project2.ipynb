```python
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
employee = pd.read_csv('tbl_Employee.csv')
performans = pd.read_csv('tbl_Perf.csv') 
```

```python
employee.head()
```

```python
performans.head()
```

```python
employee.isna().sum()
```

```python
new_employee =  employee.dropna(axis = 1)
```

```python
new_employee
```

```python
new_employee.info()
```

```python
new_employee['EngDt'] =  new_employee['EngDt'].astype('datetime64')
```

```python
new_employee['EmpID'] =  new_employee['EmpID'].astype('int64')
```

```python
performans.info()
```

```python
employee_and_performs =  pd.merge(new_employee , performans , how = 'inner' , on = 'EmpID')
```

```python
employee_and_performs.head()
```

```python
max_rating =  employee_and_performs[employee_and_performs['Rating'] == 5]
```

```python
max_rating
```

```python
max_rating['DOB'].dtype
```

```python
def age(born):
    born = datetime.strptime(born, "%m/%d/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
max_rating['DOB'].to_frame()
max_rating['Age'] = max_rating['DOB'].apply(age)
```

```python
max_rating
```

```python
def gender(sex):
  if(sex == 1):
    return 'Female'
  else :
    return 'Male'
```

```python
max_rating['Gender'] =  max_rating['GenderID'].apply(gender)
```

```python
max_rating
```

```python
max_rating.reset_index(inplace= True)
```

```python
max_rating.to_csv('max_rating_employe.csv')
```
